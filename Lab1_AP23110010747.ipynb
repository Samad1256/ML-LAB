## <a name='reshape'></a>5Ô∏è‚É£ Reshaping, Transpose & Copies vs‚ÄØViews

```python
b = np.arange(8)
B = b.reshape(2,4)
B[0,0] = 99
print('b is modified:', b)
C = b.reshape(2,4).copy()
C[0,0] = -1
print('b unchanged with copy:', b)
```

```python
# üñäÔ∏è TODO: Flatten a 3‚ÄëD array into 1‚ÄëD using both `ravel` and `flatten`; observe copy vs view.
import numpy as np

# Create a 3-D array
x = np.arange(2*3*4).reshape(2, 3, 4)
print("Original x:\n", x)

# 1) ravel: returns a view when possible (no copy)
x_ravel = x.ravel()
print("\nravel output (before modifying x):\n", x_ravel)

# Modify the original array
x[0, 0, 0] = 999
print("\nAfter modifying x[0,0,0] = 999")
print("x_ravel reflects change (view when possible):\n", x_ravel)

# 2) flatten: always returns a copy
x_flatten = x.flatten()
print("\nflatten output (before modifying x again):\n", x_flatten)

# Modify the original array again
x[0, 0, 1] = 888
print("\nAfter modifying x[0,0,1] = 888")
print("x_flatten does NOT change (it's a copy):\n", x_flatten)

# For clarity, show current x
print("\nCurrent x:\n", x)

```

## <a name='join'></a>6Ô∏è‚É£ Joining, Splitting, Set & Sorting Ops

```python
x = np.array([1,3,5]); y = np.array([2,4,6])
xy = np.concatenate([x,y])
print('union', np.union1d(x,y))
print('intersect', np.intersect1d(xy,[1,2,10]))
print('sorted descending', np.sort(xy)[::-1])
```

```python
# üñäÔ∏è TODO: split `xy` back into two equal halves using `np.array_split`
import numpy as np

x = np.array([1,3,5])
y = np.array([2,4,6])

xy = np.concatenate([x, y])

# Split into two equal halves
left, right = np.array_split(xy, 2)

print("left half:", left)    # [1 3 5]
print("right half:", right)  # [2 4 6]
```

## <a name='broadcast'></a>8Ô∏è‚É£ Broadcasting Rules

Rules: compare dimensions from right ‚Üí left; stretch size 1 dims; mismatch error.

```python
row = np.arange(5)
col = np.arange(3).reshape(3,1)
matrix = row + col  # broadcast to 3√ó5
print(matrix)
```

```python
# üñäÔ∏è TODO: use broadcasting to create a 10√ó10 multiplication table.
import numpy as np

# Create a row vector [1..10] and a column vector [1..10]
row = np.arange(1, 11).reshape(1, 10)   # shape (1,10)
col = np.arange(1, 11).reshape(10, 1)   # shape (10,1)

# Broadcasting multiplies each row element by each column element ‚Üí (10,10)
mult_table = row * col

print(mult_table)

```

```python
# üñäÔ∏è TODO: add a new field 'height' to the structured array using `np.lib.recfunctions.append_fields` (hint: pip install?).
import numpy as np

# Original structured array
people = np.array(
    [('Alice', 25, 55.0), ('Bob', 30, 85.5)],
    dtype=[('name','U10'), ('age','i4'), ('weight','f4')]
)

# If recfunctions import fails in your environment, ensure NumPy is installed/updated
# !pip install --upgrade numpy

from numpy.lib import recfunctions as rfn

# New height data (must match array length)
heights = np.array([165.0, 180.0], dtype='f4')  # centimeters, for example

# Append the new field 'height'
people_with_height = rfn.append_fields(
    people,
    names='height',
    data=heights,
    dtypes='f4',
    usemask=False
)

print(people_with_height)
print("fields:", people_with_height.dtype.names)
print("heights:", people_with_height['height'])
```

## <a name='mask'></a>1Ô∏è‚É£5Ô∏è‚É£ Masked Arrays & NaNs

```python
arr = np.array([1, 2, np.nan, 4, np.nan])
masked = np.ma.masked_invalid(arr)
print(masked.mean())
```

```python
import numpy as np

# Example 2-D array with NaNs
A = np.array([
    [1.0,  2.0,  np.nan],
    [4.0,  np.nan, 6.0],
    [np.nan, 8.0,  9.0]
], dtype=float)

# Compute column means ignoring NaNs
col_means = np.nanmean(A, axis=0)

# Create a mask of NaN positions
mask = np.isnan(A)

# Replace NaNs with the corresponding column means
A[mask] = np.take(col_means, np.where(mask)[1])

print("Column means:", col_means)
print("Filled array:\n", A)
```

```python
import numpy as np

fitness = np.genfromtxt('fitness.txt', delimiter='\t', dtype=None, encoding=None, names=True)
print('columns:', fitness.dtype.names, 'rows:', len(fitness))
```

```python
import numpy as np
import pandas as pd

fitness = np.genfromtxt('fitness.txt', delimiter=None, dtype=None, encoding=None, names=True)

cols = [c.lstrip('#') for c in fitness.dtype.names]
fitness.dtype.names = tuple(cols)

df = pd.DataFrame(fitness)

df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
df['month'] = df['date'].dt.to_period('M')
df['week'] = df['date'].dt.isocalendar().week

monthly_steps = df.groupby('month')['step_count'].sum()
sleep_mood_corr = df['hours_of_sleep'].corr(df['mood'])
weekly_summary = df.groupby('week').agg({
    'step_count': 'sum',
    'hours_of_sleep': 'mean',
    'mood': 'mean',
    'calories_burned': 'sum'
})

print('Monthly Step Count:\n', monthly_steps)
print('\nSleep vs Mood Correlation:', sleep_mood_corr)
print('\nWeekly Summary:\n', weekly_summary)
```
